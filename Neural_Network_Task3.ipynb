```
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split as splt
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib as plt
```

```
data = load_breast_cancer()
data
```

```
import pandas as pd
df = pd.DataFrame(data.data, columns=data.feature_names)
df.shape
```

```
x,y = data.data, data.target
x_train,x_test,y_train,y_test = splt(x,y, test_size=0.33, random_state=42)
```

```
from keras.layers import Dense
from keras.losses import mse,binary_crossentropy
from keras.optimizers import Adam
```

```
model = keras.Sequential()
model.add(Dense(4, activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(128, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(64, activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(optimizer = 'adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train,y_train,epochs=200)
```

```
a ,acc_train= model.evaluate(x_train,y_train)
b ,acc_test= model.evaluate(x_test,y_test)
print("Train data accuracy = {:f}".format(acc_train*100))
print("Test data accuracy = {:f}".format(acc_test*100))
```
