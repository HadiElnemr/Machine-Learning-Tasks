```
import pandas as pd
data = pd.read_csv('train.csv',index_col='PassengerId')
y = data['Survived']
data=data.drop(['Ticket','Name','Survived','Cabin'], axis=1)
data.Age = data.Age.fillna(data.Age.mean())
data["Sex"]=data["Sex"].replace({"male":str(1),"female":str(0)}).astype(int)
data['Embarked'].fillna("S",inplace=True) 
data.head()
```

```
from tensorflow import keras
from sklearn.model_selection import train_test_split as splt
import numpy as np 
```

```
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["Embarked"]=le.fit_transform(data["Embarked"])
X = data
X.Fare = X.Fare.fillna(X.Fare.mean())
```

```
from keras.layers import Dense
from keras.losses import mse,binary_crossentropy
from keras.optimizers import Adam
```

```
x_train,x_test,y_train,y_test = splt(X,y,test_size = 0.33,random_state = 42)
```

```
model = keras.Sequential()
model.add(Dense(16,activation = 'relu'))
model.add(Dense(64,activation = 'relu'))
model.add(Dense(128,activation = 'relu'))
model.add(Dense(64,activation = 'relu'))
model.add(Dense(32,activation = 'relu'))
model.add(Dense(16,activation = 'relu'))
model.add(Dense(4,activation = 'relu'))
model.add(Dense(1,activation = 'sigmoid'))
model.compile(optimizer = 'adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(x_train,y_train,epochs=200)
```

```
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=9, random_state=42)
clf.fit(x_train, y_train)
clf.score(x_test,y_test)
print(clf.predict(x_test))
```

```
model.evaluate(x_test,y_test)
```

```
pdata = pd.read_csv('test.csv',index_col="PassengerId")
pdata['Age'].fillna(pdata['Age'].mean(), inplace = True)
pdata['Cabin'].fillna("C", inplace = True)
pdata['Fare'].fillna(pdata['Fare'].mean(), inplace = True)

pdata["Sex"]=pdata["Sex"].replace({"male":str(1),"female":str(0)}).astype(int)
pdata["Embarked"]=le.fit_transform(pdata["Embarked"])
pdata.drop(["Ticket","Name",'Cabin'],inplace=True, axis=1)

# res = pd.DataFrame(model.predict(pdata),columns=['Survived'])
res = pd.DataFrame(clf.predict(pdata),columns=['Survived'])
res['PassengerId'] = pdata.reset_index()['PassengerId']
res = res.set_index('PassengerId')
res[res['Survived']>=0.5] = 1
res[res['Survived']<0.5] = 0
res = res.astype(int)
res.to_csv('output.csv')
res
```
